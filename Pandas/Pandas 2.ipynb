```python
student = ['Allwin','Hyden','Sriram','Yathishwar']
marks = [89,95,96,99]
list_of_student = list(zip(student,marks))
mark_statement = pd.DataFrame(list_of_student,columns=['Name','Marks'])
print(mark_statement)
```

```python
student_name = ['Allwin','Hyden','Sriram','Yathishwar']
subject = ['Maths','Physics','Chemistry']
maths_marks = [95,96,97,99]
physics_marks = [88,85,96,97]
chemistry_marks = [89,95,94,99]
list_of_student = list(zip(student_name,subject,maths_marks,physics_marks,chemistry_marks))
mark_statement = pd.DataFrame(list_of_student,columns = ['Name','Subject','Maths marks','Physics marks','Chemistry marks'])
print(mark_statement)
```

```python
student_marks = {'Name':['Allwin','Hyden','Yathishwar'],'Maths mark':[95,96,97],'Physics mark':[88,85,96],'Chemistry mark':[89,95,94]}
marks_statement = pd.DataFrame(student_marks)
print(marks_statement)
```
