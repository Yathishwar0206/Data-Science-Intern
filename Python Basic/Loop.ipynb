```python
# list comprehension
list_of_number=[1,2,4,6,8,11,14,15,19,20]
squared_number=[numb*2 for numb in list_of_number]
print(f'The Squared number is : {squared_number}')
```

```python
# list comprehension even
even_squared_number=[numb**2 for numb in list_of_number if numb%2==0]
print(f'The Even squared number is : {even_squared_number}')
```

```python
# list comprehension odd
odd_squared_number=[numb**2 for numb in list_of_number if numb%2!=0]
print(f'The Odd squared number is : {odd_squared_number}')
```

```python
# List comprehension of odd(if,else) in single list
odd_squared_number=[numb**2 if numb%2!=0 else numb**3 for numb in list_of_number]
print(f'The Odd squared number is : {odd_squared_number}')
```

```python
# Dictionary comprehension
vegetable_color=[('Tomato','Red'),('Pumpkin','Orange'),('Chilli','Green'),('Raddish','White')]
vegetable_color_dict=dict(vegetable_color)
print(vegetable_color_dict)
```

```python
# loop
flipped={}
vegetable_color={'Tomato':'Red','Pumpkin':'Orange','Chilli':'Green','Raddish':'White'}
for vegetable,color in vegetable_color.items():
  flipped[color]=vegetable
print(flipped)
```

```python
flipped={color:vegetable for vegetable,color in vegetable_color_dict.items() if vegetable[0] not in ['P','R']}
print(flipped)
```

```python
flipped={color[:3]:vegetable[-2:] for vegetable,color in vegetable_color_dict.items()}
print(flipped)
```

```python
# Dictionary comprehension
fruit_place=[('Apple','Kashmir','Mango','Salem','Jackfruit','Panruti','Littchi','Ooty')]
fruit_place_dict=dict(fruit_place)
print(fruit_place_dict)
```

```python
# loop
flipped={}
fruit_place={'Apple':'Kashmir','Mango':'Salem','Jackfruit':'Panruti','Littchi':'Ooty'}
for fruit,place in fruit_place.items():
  flipped[place]=fruit
print(flipped)
```

```python
flipped={place:fruit for fruit,place in fruit_place_dict.items() if fruit[0] not in ['L']}
print(flipped)
```

```python
flipped={place[:3]:fruit[-3:] for fruit,place in fruit_place_dict.items()}
print(flipped)
```
