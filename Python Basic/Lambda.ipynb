```python
# Square in Lambda function
square = lambda x: x**2
numbers= [5,7,9,11,13,15]
squared_numbers = list(map(square,numbers))
print(squared_numbers)
```

```python
# if,else
function = lambda x: 'Even' if x%2==0 else 'Odd'
print(function(2))
```

```python
# lambda map(),sorted()
student = (['yathish',95],['Yathish',85],['Ajith',100])
sorted_student = lambda student:sorted(student,key=lambda student:student[1])
print(sorted_student[1])
```
