```python
import matplotlib.pyplot as plt
import seaborn as sns
level = [15,25,40,20]
label = ['A','B','C','D']
color = sns.color_palette('colorblind',len(level))
plt.pie(level,labels=label,colors=color,autopct='%1.2f%%',startangle=90)
plt.show()
```

```python
# Company sales
import pandas as pd
company_data = pd.read_csv(r'C:\Users\ADMIN\Desktop\Jupyter folder/Downloads/company_sales_data.csv')
print(company_data)
```

```python
company_sales = company_data['total_profit'].sum()
print(company_sales)
```

```python
company_yearly_sales = company_data[['facecream','facewash','toothpaste','bathingsoap','shampoo','moisturizer']].sum()
print(company_yearly_sales)
```

```python
company_yearly_sales = company_data[['facecream','facewash','toothpaste','bathingsoap','shampoo','moisturizer']].sum()
# company_month = pd.DataFrame(company_data)
# company_month['month_number'] = pd.to_datetime(company_month['month_number'],format='%m'),company_month.strftime('%b')
print(company_yearly_sales)
```

```python
# Line chart
plt.figure(figsize=(10,5))
plt.plot(company_data['month_number'],company_data['facecream'],marker='o',label ='Face cream',linestyle='--')
plt.plot(company_data['month_number'],company_data['facewash'],marker='o',label ='Face wash',linestyle='-')
plt.plot(company_data['month_number'],company_data['toothpaste'],marker='o',label ='Tooth paste',linestyle='--')
plt.plot(company_data['month_number'],company_data['bathingsoap'],marker='o',label ='Bathing soap',linestyle='-')
plt.plot(company_data['month_number'],company_data['shampoo'],marker='o',label ='Shampoo',linestyle='--')
plt.plot(company_data['month_number'],company_data['moisturizer'],marker='o',label ='Moisturizer',linestyle='-')
plt.xlabel('Month')
plt.ylabel('Sold')
plt.title('Product sales')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Bar chart
company_yearly_sales.plot(kind='bar', figsize=(10,5), color='green')
plt.title('Company yearly sales of each product')
plt.ylabel('Sales')
plt.xlabel('Product')
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```

```python
company_yearly_sales.plot(kind='pie', figsize=(10,5), color='red',autopct='%1.1f%%')
plt.title('Company yearly sales of each product')
plt.tight_layout()
plt.show()
```
