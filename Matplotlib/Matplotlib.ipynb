```python
import matplotlib.pyplot as plt
import pandas as pd
# Show the first 5 rows of the DataFrame.
student_data = pd.read_csv(r'/content/students_20.csv')
print(student_data)
print(student_data.head(5))
```

```python
# Show only the "Student Name" and "Math" columns.
print(student_data[['Student Name','Math']])
```

```python
# Find the average marks in each subject.
print(student_data[['Math','Science','English']].mean())
```

```python
# Which student scored the highest in Science?.
print(student_data.loc[student_data['Science'].idxmax(),'Student Name'])
```

```python
# Select students who scored more than 90 in English.
print(student_data[student_data['English']>90 ])
```

```python
# Sort students by Math marks in descending order.
print(student_data.sort_values('Math',ascending=False))
```

```python
# Add a new column "Total" = Math + Science + English.
student_data['Total'] = student_data['Math']+student_data['Science']+student_data['English']
print(student_data)
```

```python
# Find the student with the highest total marks.
print(student_data.loc[student_data['Total'].idxmax(),['Student Name','Total']])
```

```python
# Show students who scored less than 80 in at least one subject.
print(student_data[(student_data['Math'] < 80) | (student_data['Science'] < 80) | (student_data['English'] < 80 )])
```

```python
print(student_data[(student_data['Math']<80) | (student_data['Science']<80 )| (student_data['English'] < 80)])
```

```python
# Save the updated DataFrame (with totals) to a new CSV.
student_data.to_csv('Student_data_with_total.csv')
print(student_data)
```

```python
# What is the highest and lowest mark in Math?.
print('Highest mark : ',student_data['Math'].max())
print('Lowest mark : ',student_data['Math'].min())
```

```python
# Find the average total marks of all students.
print('Average Mark : ',student_data['Total'].mean())
```

```python
# Show only the top 3 students in English.
print(student_data.sort_values('English',ascending=False).head(3))
```

```python
# Count how many students scored above 85 in Science.
print('No of students got above 80 in Science : ',(student_data['Science']> 80).sum())
```

```python
# Create a new column "Grade" based on Total Marks.
def grade(total):
  if total >= 250:
    return 'A'
  elif total >= 200:
    return 'B'
  elif total >= 150:
    return 'C'
  else:
    return 'D'
student_data['Grade'] = student_data['Total'].apply(grade)
print(student_data)
```

```python
# Show the average marks per subject for each Grade.
print(student_data.groupby('Grade')[['Math','Science','English']].mean())
```

```python
# Show students who have the same marks in any two subjects.
print(student_data[(student_data['Math'] == student_data['Science']) |
 (student_data['Science'] == student_data['English']) |
   (student_data['English']==student_data['Math'])])
```

```python
# Show the top 5 students based on total marks only (Student Name + Total).
print(student_data[['Student Name','Total']].sort_values('Total',ascending=False).head(5))
```

```python
# Save only Student Name and Grade to a new CSV.
student_data[['Student Name','Grade']].to_csv('student_grades.csv')
print(student_data)
```
