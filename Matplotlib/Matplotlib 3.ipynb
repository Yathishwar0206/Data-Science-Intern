```python
import matplotlib.pyplot as plt
import pandas as pd
student_data = pd.read_csv(r'C:\Users\ADMIN\Desktop\Jupyter folder/Downloads/student_marks.csv')
print(student_data)
```

```python
# Total marks
student_data['Total marks'] = student_data['Math']+student_data['Science']+student_data['English']
student_total_marks = student_data['Total marks']
print(student_total_marks)
```

```python
student_total = student_data[['Math','Science','English']].sum()
print('Total Marks for per subject : ',student_total)
```

```python
# Line chart
plt.figure(figsize=(10,5))
plt.plot(student_data['Math'],marker='o',linestyle='--',label = 'Maths')
plt.plot(student_data['Science'],marker='o',linestyle='--',label = 'Science')
plt.plot(student_data['English'],marker='o',linestyle='--',label = 'English')
plt.xlabel('Student')
plt.ylabel('Subject marks')
plt.title('Students marks')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Bar chart
student_total.plot(kind='bar', figsize=(10,4))
plt.xlabel('Student')
plt.ylabel('Subject marks')
plt.title('Student marks')
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```

```python
# Pie chart
student_total.plot(kind='pie', figsize=(10,4), autopct='%1.1f%%',startangle= 120)
plt.title('Student marks')
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```

```python
student_total.plot(kind='barh', figsize=(10,4),)
plt.ylabel('Student')
plt.xlabel('Subject marks')
plt.title('Student marks')
plt.grid(axis='x')
plt.tight_layout()
plt.show()
```

```python
# Top 10 students by their total marks.
student_data['Total'] = student_data['Math']+student_data['Science']+student_data['English']
Top10 = student_data.nlargest(10,'Total')
plt.bar(Top10['Student'],Top10['Total'],color='Green')
plt.title('Top 10 students by their marks')
plt.ylabel('Marks')
plt.show()
```

```python
# Top 10 weak students by their students.
student_data['Total'] = student_data['Math']+student_data['Science']+student_data['English']
Weak10 = student_data.nsmallest(10,'Total')
plt.bar(Weak10['Student'],Weak10['Total'],color='Red')
plt.title('10 Weak students by their marks')
plt.ylabel('Total marks')
plt.show()
```

```python
plt.figure(figsize=(10,5))
plt.hist(student_data, bins=50)
plt.title('Student Total marks')
plt.show()
```
