## Visualize the fuel type of the car using bar chart.

```python
plt.figure(figsize=(15,9))
color = ['red','orange','blue','brown','green']
edgecolor = ['black','red','green','red','lightgreen']
df['fuelType'].value_counts().plot(kind='bar',color=color,edgecolor=edgecolor)
plt.title('Fueltype of the car', fontweight='bold')
plt.xlabel('Fueltype', fontweight='bold')
plt.ylabel('Counts',fontweight='bold')
plt.tight_layout()
plt.show()
```

## Observation
- The above dataset, **Diesel** type car were sold highly.
- **Electric** vehicle got **Lower**.

## Transmission

```python
gear_type = df.groupby('model')['transmission'].value_counts()
gear_type
```

```python
plt.figure(figsize=(15,9))
color = ['red','green','yellow']
gear_type.plot(kind='bar',color=color,edgecolor='black')
plt.title('Gear type of the Car',fontweight='bold')
plt.xlabel('Transmission',fontweight='bold')
plt.ylabel('Count',fontweight='bold')
plt.tight_layout()
for index,value in enumerate(gear_type):
    plt.text(index, value+5, str(value), ha='center', va='bottom', fontweight='bold', fontsize=6)
plt.show()
```

## Observation
- The Above dataset, **1103** cars sold mostly.In the transmission of **Manual** and the model is **BMW Series 1**.
- **BMW M2** got **Lower**.

## Model and Year

```python
model_year = df.groupby('year')['model'].apply(list)
for year, model in model_year.items():
    print(f'\n Year : {year}')
    print(f'\n Model : {model}')
```

```python
model_per_year = df.groupby('year')['model'].count().reset_index()
plt.figure(figsize=(10,5))
plt.bar(model_per_year['year'], model_per_year['model'], color='green')
plt.xlabel('Year',fontweight='bold')
plt.ylabel('Model',fontweight='bold')
plt.title('Number of Model per Year',fontweight='bold')
plt.tight_layout()
plt.show()
```

## Observation
- In the year, **2019** most of the car were sold in that year.

## Model vs Year vs Mileage

```python
sns.scatterplot(x='model', y='year',hue='mileage', data=df)
plt.title('Mileage of the car and the Year',fontweight='bold')
plt.xticks(rotation=90)
plt.show()
```

## Observation
- According to above dataset, In the year **2020** model car were run high mileage than other cars.

## Model vs Engine size

```python
ME = df.groupby('model')['engineSize'].value_counts()
print(ME)
```

```python
plt.figure(figsize=(10,6))
ME.plot(kind='bar', color='skyblue', edgecolor='blue')
plt.title('Model vs Engine size',fontweight='bold')
plt.xlabel('Model',fontweight='bold')
plt.ylabel('Engine size',fontweight='bold')
plt.tight_layout()
for index, value in enumerate(ME):
    plt.text(index, value+5, str(value), ha='center', va='bottom', fontweight='bold', fontsize=7)
plt.show()
```

## Observation
- The Above dataset,In **BMW Series 3** mostly the engine size is **2.0**. It is got higher.
