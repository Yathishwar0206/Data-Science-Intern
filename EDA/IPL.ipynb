## Handle missing values appropriately:
Fill missing values in player_dismissed, dismissal_kind, and fielder in the deliveries dataset.

```python
deliveries['player_dismissed'].fillna(value='ban',inplace=True)
print(deliveries['player_dismissed'].isnull().sum())
```

```python
deliveries['dismissal_kind'].fillna(value='dismissed',inplace=True)
print(deliveries['dismissal_kind'].isnull().sum())
```

```python
deliveries['fielder'].fillna(value='fielder',inplace=True)
print(deliveries['fielder'].isnull().sum())
```

```python
matches['city'].fillna(value='city',inplace=True)
print(matches['city'].isnull().sum())
```

```python
matches['player_of_match'].fillna(value='man of the match',inplace=True)
print(matches['player_of_match'].isnull().sum())
```

```python
matches['winner'].fillna(value='winner',inplace=True)
print(matches['winner'].isnull().sum())
```

```python
matches['umpire2'].fillna(value='umpire2',inplace=True)
print(matches['umpire2'].isnull().sum())
```

```python
matches['umpire1'].fillna(value='umpire1',inplace=True)
print(matches['umpire1'].isnull().sum())
```

Drop the umpire3 column since it has all missing values

```python
matches.drop(columns=['umpire3'], axis=1, inplace=True)
```

## Observation
1. In **Deliveries**, I filled the datas using **fillna()** method. The columns **Fielder, Player Dismissed, Dismissed kind** are replaced.
2. In **Matches**, The same like the deliveries. Columns **Umpire1. Umpire2, City, Winner, Player of Match** are replaced
3. The column **Umpire3** is Dropped because the 80-90% of the datas are missing. So, it is better to **Drop** the column.

## Display the champion team of each IPL season(the winner of the final match) and visulize the number of titles won by each team using bar chart.

## Calculate the percentage of matches won by teams batting second and visualize the comparison of wins(batting first vs batting second).

```python
bat_second_wins = matches[matches['win_by_wickets'] > 0].shape[0]
total_matches = matches[matches['result']=='normal'].shape[0]
percent_bat_second = round(bat_second_wins/total_matches*100,2)
plt.bar(['Wins batting 2nd','Other wins'],[bat_second_wins, total_matches-bat_second_wins], color = ['blue','grey'])
plt.title(f"Percentage wins batting second : {percent_bat_second}")
plt.xlabel('Wins type')
plt.ylabel('Matches')
plt.tight_layout()
plt.show()
```

## Observation
- According to the above dataset, **Batting second** has **54.15%** highest chance to win.

## Identify the bowlers who have bowled the most number of balls in IPL and Visualize the top 10.
